# Percentual de Faturamento


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams.update({'font.size':20})
```

```python
cidade = ['SP', 'RJ', 'MG', 'ES', 'OUTROS']
faturamento = [67.83643, 36.67866, 29.22988, 27.16548, 19.84953 ]
```

```python
zip(cidade, faturamento)
```

```python
list(zip(cidade, faturamento))
```

```python
explode = (0.1, 0, 0, 0, 0)
```

```python
plt.figure(figsize=(7,7))
plt.pie(x=faturamento, labels=cidade, explode=explode, autopct='%1.1f%%', startangle=90)

plt.show()
```
